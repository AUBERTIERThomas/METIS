### ``1`` Change date

While this does not interfere greatly with the main process, it can be convenient still.

``geophpy.change_date`` :
* File or dataframe list (or a single one)
* New date (format is **mm/dd/yyyy**)
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/CMD mini explorer 3L GPS/HCP/")

geo.change_date("hcpmini-Tr.dat","07/15/2018",in_file=False)
```

### ``2`` Remove wrong column labels

Some devices can add column labels even when there is no column associated.
That means that every following label is shifted to the right.

``geophpy.pop_and_dec`` remove those labels while not interfering with data :
* File or dataframe list (or a single one)
* Column label to remove
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

*Note : ``geophpy.init_process`` and some other functions are checking for unused ``"Time"`` and ``"Date"`` columns automatically with ``geophpy.check_time_date``.*

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/donnees SG/2018_CMDm/2018_CMDm_zone1_test/")

pd.read_csv("Pas18_CMDm_sq1_B1.dat", sep=',')
```

```python
geo.pop_and_dec("Pas18_CMDm_sq1_B1.dat","Time",sep=',',in_file=False)
```

### ``3`` Switch columns

This may be necessary if ``"Easting"`` and ``"Northing"`` are inversed for instance.

``geophpy.switch_cols`` switch two label position while not interfering with data :
* File or dataframe list (or a single one)
* Column label A
* Column label B
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

*Note : ``geophpy.init_process`` is checking for switched ``"Easting"`` and ``"Northing"`` columns automatically.*

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/CMD mini explorer 3L GPS/HCP/")

pd.read_csv("hcpmini-Tr.dat", sep='\t')
```

```python
geo.switch_cols("hcpmini-Tr.dat","Easting","Northing")
```

### ``4`` Remove columns

The following function does only apply to unloaded file, because it would be similar to ``pandas.drop`` or ``pandas.filter``.

``geophpy.remove_cols`` remove specified columns :
* File or dataframe list (or a single one)
* List of column names to remove
* ``keep`` If the specified columns are to be kept, and removing the others instead (default False)
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

*Note : For a more automatic yet less flexible approach, see ``geophpy.light_format``.*

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/CMD mini explorer 3L GPS/HCP/")

pd.read_csv("hcpmini-Tr.dat", sep='\t')
```

```python
geo.remove_cols("hcpmini-Tr.dat",["DOP","Satelites"])
```

### ``5`` Remove areas of data

Some errors can produce wrong data while not being NaN, like divergent positions.

Forcing those to be NaN can allows NaN completion to overwrite them.
Also, leaving buggy data can mess up plotting.

``geophpy.remove_data`` empty a specific block of line on a specific range of columns :
* File or dataframe list (or a single one)
* List of column names to consider
* Index of first line of the block
* Index of last line of the block
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

*Note : ``geophpy.stats`` can help localize anomalies by showing line indexes of extreme values.*

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/donnees SG/2018_CMDm/2018_CMDm_zone1_test/")

pd.read_csv("Pas18_CMDm_sq8.dat", sep=',')
```

```python
sq1 = geo.pop_and_dec("Pas18_CMDm_sq8.dat","Time",sep=',',in_file=False)[0]

geo.data_stats(sq1,["x[m]","y[m]"],sep=',',bins=15,n=20)
```

```python
sq1 = geo.remove_data(sq1,["x[m]","y[m]"],2268,2283,sep=',',in_file=False)[0]
print(sq1.loc[2267:2284,["x[m]","y[m]"]])
```

### ``6`` Lighten dataset

This procedure remove less used data column as well as putting positions and data at the same indexes.

``geophpy.light_format`` :
* File or dataframe list (or a single one)
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``nb_paths`` Number of paths (default 3)
* ``restr`` Exclusion strings : any data including one of the specified strings will be ignored
* ``in_file`` Return the dataframe list instead if False (default False)

The output column order is shown below.

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/CMD mini explorer 3L GPS/HCP/")

app_data = gemi.add_device('mini3L','HCP',3,[30000],gps=True,gps_dec=[0.25,-0.2],TR_l=[0.32,0.71,1.18],
                           TR_t=[0,0,0],height=0.1,bucking_coil=0,
                           coeff_c_ph=[0.5982,0.9839,0.7074],coeff_c_qu=[0.00599,0.0290,0.0785],save=True)
hcp_b, hcp_p = gemi.init_process(app_data,["hcpmini-Tr.dat","hcpmini-Tr2.dat","hcpmini-Tr3.dat"],
                                 sep='\t',sup_na=False,regr=False,corr_base=False,plot=False)
```

```python
geo.light_format(hcp_p[0],restr=["Inv"])
```

### ``8`` Set position for no GPS data

Raw files may not have a set Y position. In this case, the file should have a special line between each profile indication the final position of the profile before.

Profile getting negative X position will be considered as a base.

``geophpy.no_gps_pos`` :
* File or dataframe list (or a single one)
* ``sep`` Dataframe separator (default ``'\t'``)
* ``replace`` If input file is overwritten (default False)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/Divers/")

pd.read_csv("Lim2705Lo.dat", sep='\t')[:10]
```

```python
geo.no_gps_pos("Lim2705Lo.dat", sep='\t')[0][:10]
```

### ``9`` Manage external bases

This function assumes that one profile prospection file has its before and after bases in two separate files.

``geophpy.fuse_bases`` :
* First base file
* Second base file
* Profile file
* ``sep`` Dataframe separator (default ``'\t'``)
* ``output_file_list`` List of output file name
* ``in_file`` Return the dataframe list instead if False (default False)

Output can then be used as any other base dataframe that is extracted with the classical processes.

```python
# Your data path, if not in current working directory
os.chdir("/home/taubertier/StageMETIS/2024 2025 Aubertier/fichiers donnees/donnees SG/2018_CMDm/2018_CMDm_zone1_test/")

app_data = gemi.add_device('mini3L','HCP',3,[30000],gps=False,gps_dec=[0.25,-0.2],TR_l=[0.32,0.71,1.18],
                           TR_t=[0,0,0],height=0.1,bucking_coil=0,
                           coeff_c_ph=[0.5982,0.9839,0.7074],coeff_c_qu=[0.00599,0.0290,0.0785],save=True)
gemi.init_process(app_data,["Pas18_CMDm_sq1.dat"],sep=',',
                  sup_na=False,regr=False,corr_base=False,plot=False,in_file=True)
```

```python
base = gemi.fuse_bases("Pas18_CMDm_sq1_B1.dat","Pas18_CMDm_sq1_B2.dat","Pas18_CMDm_sq1_init_P.dat", sep=',')
print(base)
```
